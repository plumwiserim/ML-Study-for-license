```python
# 1. Insurance 고객 다중 분류 모형을 구축하고자 한다. 
# Segmentation 은 1, 2, 3, 4로 각 고객 유형을 분류한 후 
# Macro-F1 Score로 평가하고 
# 4th_insurance_test 평가 데이터셋으로부터 산출한 y_predict_result를 제출하기

df_train = pd.read_csv('data\\4th_insurance.csv') 
df_test = pd.read_csv('data\\4th_insurance_test.csv')
```

```python
df_train.info()
```

```python
df_train.head()
```

```python
x_data = df_train.iloc[:, :-1]
y_data = df_train['Segmentation']
```

```python
x_data = pd.get_dummies(x_data)
x_data
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# min/max scaler
# robust scaler
# 지수, 로그 변환.. 
# 전처리 스케일러 필터링 등 

x_train, x_valid, y_train, y_valid = train_test_split(x_data, y_data, test_size = 0.3, random_state = 11)
print(x_train.shape)
print(x_valid.shape)
print(y_train.shape)
print(y_valid.shape)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
y_pred = rf_model.predict(x_valid)
accuracy = accuracy_score(y_valid, y_pred)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score

# 다중 분류 모델이기 때문에 (Segmentaiton 1, 2, 3, 4)
# 4 x 4 matrix
confusion_matrix(y_valid, y_pred)
```

```python
# 불균형 데이터의 경우, accuracy 이 적절한 평가지표가 아니기 때문에
# accuracy 보다 f1_score 가 더 신뢰적이다. 
f1_score(y_valid, y_pred, average='macro')  # average = 'micro'4.ipynb
```

```python
# gridsearchcv 를 통해서 model selection 
# gridsearchcv 는 시간이 오래 걸린다는 단점이 있지만 더 좋은 파라미터를 찾아줄 수 있다. 
params = {
    'n_estimators': [10, 50, 75, 100], 
    'max_depth': [3, 6, 8, 10, 12], 
    'max_features': [2, 4, 6, 8]
    # min_samples_leaf
    # min_samples_split
}

grid_search = GridSearchCV(rf_model, params, cv = 5) 
grid_search.fit(x_train, y_train)

print('최적의 Parameter: ', grid_search.best_params_)

# 후보 hyper parameter 선정 
 
# cross-validation 을 통해 좋은 파라미터 선택 

```

```python
rf_model = RandomForestClassifier(
    max_depth = 8, max_features = 8, n_estimators = 75
)
rf_model.fit(x_train, y_train)
y_pred = rf_model.predict(x_valid)
f1_score(y_valid, y_pred, average='macro')
```

```python
df_test = pd.read_csv('data\\4th_insurance_test.csv')
```

```python
df_test = pd.get_dummies(df_test)
y_result = rf_model.predict(df_test)

df_test['predict_result'] = y_result
```
