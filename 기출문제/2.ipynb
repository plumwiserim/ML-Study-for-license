```python
# 2. 주어진 데이터셋에서 feature2 상위 80%인 학습 데이터셋 생성한 후, 
# Feature1 의 결측치를 Feature1의 중앙값으로 대체하고, 
# 대체하기 전 후의 표준편차 차이 구하기 

df = pd.read_csv('data\\case1_no2.csv')
# print(df.shape)   # (105, 2)

df = df.sort_values(by='Feature2', ascending = False).reset_index(drop=True)

# from sklearn.model_selection import train_test_split
# df_train, df_test = train_test_split(df, test_size = 0.2) # 오답 

idx = int(len(df)* 0.8)
print(idx)
old_df = df.iloc[:idx, :]   # 정답 ! 

print(old_df.shape) # (84, 2)
print(old_df['Feature1'].std())

# df_train.isnull().sum()

value = old_df['Feature1'].median()   # 오답 
value = old_df[old_df['Feature1'].notnull()]['Feature1'].median()   # 정답 ! 
print(value)

new_df = old_df.fillna(value)
new_df.isnull().sum()


old_std = old_df['Feature1'].std()
new_std = new_df['Feature1'].std()

print(old_std)
print(new_std)
print(old_std - new_std)
```

```python
# 3. feature1 의 이상치를 구하여 이상치들의 합계 구하기 
# (이상치는 평균으로부터 표준편차*1.5 만큼 위 아래로 벗어난 값을 의미)

df = pd.read_csv('data\\case1_no1.csv')

mean = df['Feature1'].mean()
std = df['Feature1'].std()
threshold = std * 1.5 

df[(df['Feature1'] > mean + threshold) | (df['Feature1'] < mean - threshold)]['Feature1'].sum()
```

```python
# 3. feature1 의 이상치를 구하여 이상치들의 합계 구하기 
# (이상치는 q1, q3으로부터 사분위범위*1.5 만큼 위 아래로 벗어난 값을 의미)

df = pd.read_csv('data\\case1_no1.csv')

q1 = np.percentile(df['Feature1'], 25)
q3 = np.percentile(df['Feature1'], 75)
iqr = q3 - q1
lower_limit = q1 - iqr * 1.5
upper_limit = q3 + iqr * 1.5

df[(df['Feature1'] > upper_limit) | (df['Feature1'] < lower_limit)]['Feature1'].sum()
```

```python
import pandas as pd 

df = pd.read_csv('data\\case2_no1.csv')
print(df.info())

print(df['bankrupt'].value_counts())
# --> 과도한 불균형 데이터가 아니다 

# 결측치 처리 
# print(df.isnull().sum())
replacing_value = df['score'].mean()
df['score'] = df['score'].fillna(replacing_value)
# print(df.isnull().sum())

# 이상치 처리 
print(df['Pclass'].value_counts())
df.loc[df['Pclass'] == '*3', 'Pclass'] = '3'
print(df['Pclass'].value_counts())

# 더미 변수 생성 
df = pd.get_dummies(df)

# 다른 방법 
# df['normal'] = df['grade'].apply(lambda x: 1 if x=='normal' else 0)
# df['good'] = df['grade'].apply(lambda x: 1 if x=='good' else 0)

display(df)

X = df.iloc[:, 1:]
y = df['bankrupt']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 11)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
y_pred = rf_model.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)

from sklearn.model_selection import GridSearchCV

# 하이퍼파라미터 튜닝 
params = {
    'n_estimators': [10, 50, 100], 
    'max_depth': [3, 6, 8, 10], 
    'max_features': [2, 4, 6]
}

grid_search = GridSearchCV(rf_model, params, cv = 5)
grid_search.fit(x_train, y_train)
print(grid_search.best_params_)
```
