```python
# 2. Disease 2000년도 전체 국가의 유병률 평균을 구하고, 
# 2000년도에 해당 평균보다 유병률이 높은 나라의 개수 구하기 

df = pd.read_csv('data\\3rd_disease.csv')
```

```python
# 2. Disease 2000년도 전체 국가의 유병률 평균을 구하고, 
# 2000년도에 해당 평균보다 유병률이 높은 나라의 개수 구하기 

df.set_index('year', drop=True, inplace=True)
display(df)

values_of_2000 = df[df.index==2000].values
mean = values_of_2000.mean()
print(mean)
```

```python
count = 0 

for i in values_of_2000[0]: 
    if i > mean : 
        count += 1

print(count)
```

```python
# 3. Titanic 모든 컬럼 중 결측치 비율이 가장 높은 변수명 구하기 

df = pd.read_csv('data\\3rd_titanic.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# 3. Titanic 모든 컬럼 중 결측치 비율이 가장 높은 변수명 구하기 

null_rate = pd.DataFrame(df.isnull().sum() / len(df))

null_rate
```

```python
result = null_rate[null_rate[0] == max(df.isnull().sum() / len(df))].index[0]
print(result)
```

```python
# 다른 풀이 
result = df.isnull().sum().sort_values(ascending=False).index[0]
print(result)
```

```python
# Travel insurance prediction data 
# 여행객의 정보를 기반으로 여행보험 상품 가입여부를 예측하는 문제 
# 주의: index 컬럼은 독립변수로 사용하지 말 것 
```

```python
import pandas as pd 

df_train = pd.read_csv('data\\3rd_travel.csv')
```

```python
df_train.info()
```

```python
df_train.head()
```

```python
X_data = df_train.iloc[:, :-1]
y_data = df_train['TravelInsurance']
```

```python
X_data = pd.get_dummies(X_data)
X_data
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_data, y_data, test_size = 0.3, random_state = 11)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
y_pred = rf_model.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score

confusion_matrix(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='macro')
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [10, 50, 75, 100], 
    'max_depth': [3, 6, 8, 10, 12], 
    'max_features': [2, 4, 6, 8]
}

grid_search = GridSearchCV(rf_model, params, cv = 5)
grid_search.fit(x_train, y_train)

print("최적의 hyperparameter: ", grid_search.best_params_)
```

```python
rf_model = RandomForestClassifier(
    max_depth = 6, 
    max_features = 6, 
    n_estimators = 100 
)
rf_model.fit(x_train, y_train)
y_pred = rf_model.predict(x_test)

f1_score(y_test, y_pred, average='macro')
```
