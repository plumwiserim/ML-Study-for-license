```python
# 2. Disease 2000년도 전체 국가의 유병률 평균을 구하고, 
# 2000년도에 해당 평균보다 유병률이 높은 나라의 개수 구하기 

df = pd.read_csv('data\\3rd_disease.csv')
```

```python
# 다른 풀이 1
avg = df.loc[2][1:].mean()
print(avg)

num = sum(df.loc[2][1:]>avg)
print(num)
```

```python
# 다른 풀이 2
avg = df[df['year'] == 2000].iloc[0, 1:].mean()
print(avg)

num = sum(df[df['year'] == 2000].iloc[0, 1:] >= avg)
print(num)
```

```python
# 2. Disease 2000년도 전체 국가의 유병률 평균을 구하고, 
# 2000년도에 해당 평균보다 유병률이 높은 나라의 개수 구하기 

df.set_index('year', drop=True, inplace=True)
display(df)

values_of_2000 = df[df.index==2000].values
mean = values_of_2000.mean()
print(mean)
```

```python
count = 0 

for i in values_of_2000[0]: 
    if i > mean : 
        count += 1

print(count)
```

```python
# 3. Titanic 모든 컬럼 중 결측치 비율이 가장 높은 변수명 구하기 

df = pd.read_csv('data\\3rd_titanic.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# 3. Titanic 모든 컬럼 중 결측치 비율이 가장 높은 변수명 구하기 

null_rate = pd.DataFrame(df.isnull().sum() / len(df))

null_rate
```

```python
result = null_rate[null_rate[0] == max(df.isnull().sum() / len(df))].index[0]
print(result)
```

```python
# 다른 풀이 
result = df.isnull().sum().sort_values(ascending=False).index[0]
print(result)
```
