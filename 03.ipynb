## 03. Grouping

```python
import pandas as pd

# Q44. 데이터를 로드하고 상위 5개 컬럼을 출력하라
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df.head(5)
```

```python
# Q45. 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라
df['host_name'].value_counts().sort_index().head(5)
```

```python
# Q46. 데이터의 각 host_name 의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라 
# 빈도수 컬럼은 counts로 명명하라 
# ans = df['host_name'].value_counts().sort_values(ascending=False)
# ans = pd.DataFrame(ans)
# ans.columns = ['counts']
# ans.reset_index(inplace=True)
# ans.columns = ['host_name', 'counts']
# ans.set_index('host_name', drop=True, inplace=True)
# ans.head(5)

ans = df.groupby('host_name').size().to_frame().rename(columns={0:'counts'}).sort_values('counts', ascending=False)
ans.head(5)
```

```python
# Q47. neighbourhood_group의 값에 따른 neighbourhood 컬럼 갑싀 갯수를 구하여라 
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index = False).size()
```

```python
# Q48. neightbourhood_group의 값에 따른 neighbourhood 컬럼 값 중 neighbourhood_group 그룹의 최대값들을 출력하라
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index = False).size().groupby(['neighbourhood'], as_index = False).max().head()
```

```python
Ans= df.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()\
                  .groupby(['neighbourhood_group'], as_index=False).max()
Ans
```

```python
# Q49. neighbourhood_group 값에 따른 price 값의 평균, 분산, 최대, 최소 값을 구하여라 
df.groupby(['neighbourhood_group', 'price'], as_index = False).size()\
    .groupby(['price'], as_index=False).mean()
```
