### 유튜브 인기동영상 데이터

### 유튜브 공범컨텐츠 동영상 데이터

```python
import pandas as pd

channel = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

```python
# Q11. 각 데이터의 'ct'컬럼을 시간으로 인식할 수 있게 datatype을 변경하여 
# video 데이터의 videoname의 각 value마다 몇 개의 데이터씩 갖고 있는지 확인해라 
channel['ct'] = pd.to_datetime(channel['ct'])
video['ct'] = pd.to_datetime(video['ct'])

video['videoname'].value_counts()
```

```python
# Q12. 수집된 각 vide의 가장 최신화된 날짜의 viewcount값을 출력하라 
# videoname 을 기준으로 정렬, 그 다음 ct를 기준으로 정렬
# 각 videoname 별로 최신 날짜만 남기고 중복 제거 
# viewcount, videoname, ct 컬럼만 남기기 
# 인덱스 재정렬
video.sort_values(['videoname','ct'])\
    .drop_duplicates('videoname',keep='last')\
    [['viewcnt','videoname','ct']]\
    .reset_index(drop=True)
```

```python
# Q13. Channel 데이터 중 2021-10-03일 이후 각 채널의 처음 기록됐던 구독자 수(subcnt)를 출력하라 
import datetime

filtered_channel = channel[channel['ct']>=datetime.datetime(2021, 10, 3)]
filtered_channel.sort_values(['channelname', 'ct'])\
    .drop_duplicates('channelname', keep='first')\
    [['channelname', 'subcnt']]\
    .reset_index(drop=True)
```

```python
# Q14. 각 채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수(subcnt)의 증가량을 구하여라 
filtered_channel = channel[(channel['ct']>=pd.to_datetime('2021-10-03 03:00:00'))&(channel['ct']<=pd.to_datetime('2021-11-01 15:00:00'))]
first_channel = filtered_channel.sort_values(['channelname', 'ct'])\
    .drop_duplicates('channelname', keep='first')\
    [['channelname', 'subcnt']]\
    .reset_index(drop=True)
last_channel = filtered_channel.sort_values(['channelname', 'ct'])\
    .drop_duplicates('channelname', keep='last')\
    [['channelname', 'subcnt']]\
    .reset_index(drop=True)

answer = pd.DataFrame()
answer['channelname'] = first_channel['channelname']
answer['del'] = last_channel['subcnt'] - first_channel['subcnt']
answer
```

```python
# Q15. 각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요 수, 댓글 수가 수집된 것으로 알려졌다. 
# 공범 EP1의 비디오 정보 데이터 중 수집간격이 5분 이하, 20분 이상인 데이터 구간(해당 시점 전, 후)의 시각을 모두 출력하라 

# ep_one = video[video['videoname']=='공범 EP1'].sort_values('ct').reset_index(drop=True)
ep_one = video[video['videoname'].str.contains('1')].sort_values('ct').reset_index(drop=True)
ep_one 

ep_one[(ep_one['ct'].diff()<=datetime.timedelta(minutes=5)) | (ep_one['ct'].diff()>=datetime.timedelta(minutes=20))]
```

```python
# Q16. 각 에피소드의 시작날짜(년-월-일)을 에피소드 이름과 묶어 데이터프레임으로 만들고 출력하라 
video['date'] = video['ct'].dt.date

video.sort_values(['videoname', 'date']).drop_duplicates('videoname', keep='first')[['date', 'videoname']]
```

```python
# Q17. '공범' 컨텐츠의 경우 19:00시에 공개되는 것으로 알려져있다. 
# 공개된 날의 21시의 viewcnt, ct, videoname으로 구성된 데이터프레임을 viewcnt을 내림차순으로 정렬하여 출력하라 
filtered_video = video[video['ct'].dt.hour>=21]
filtered_video.sort_values(['videoname', 'ct'])\
    .drop_duplicates('videoname', keep='first')[['videoname', 'viewcnt', 'ct']].reset_index(drop=True)
```

```python
# Q18. video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 
# videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라 
filtered_video = video.sort_values(['videoname', 'ct']).drop_duplicates('videoname', keep='last')
filtered_video['ratio'] = filtered_video['dislikecnt'] / filtered_video['likecnt']
filtered_video.sort_values(by='ratio', ascending=True)[['videoname', 'ratio']].reset_index(drop=True)
```

```python
# Q19. 2021-11-01 00:00:00 ~ 15:00:00 까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오 
filtered_video = video[(video['ct']>=pd.to_datetime('2021-11-01 00:00:00'))\
                      &(video['ct']<=pd.to_datetime('2021-11-01 15:00:00'))]
first_video = filtered_video.sort_values(['videoname', 'ct'])\
    .drop_duplicates('videoname', keep='first')[['videoname', 'viewcnt']].reset_index(drop=True)
last_video = filtered_video.sort_values(['videoname', 'ct'])\
    .drop_duplicates('videoname', keep='last')[['videoname', 'viewcnt']].reset_index(drop=True)
ans = pd.DataFrame()
ans['videoname'] = first_video['videoname']
ans['viewcnt'] = last_video['viewcnt'] - first_video['viewcnt']
ans.set_index('videoname', drop=True, inplace=True)
ans
```
