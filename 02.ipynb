## 02. Filtering & Sorting

```python
import pandas as pd

# Q20. 데이터를 로드하라 
dataurl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(dataurl)
type(df)
```

```python
# Q21. quantity 컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라 
# df[df['quantity']==3].head(5)
df.loc[df['quantity']==3].head()
```

```python
# Q22. quantity 컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라 
# df[df['quantity']==3].sort_values(by=df.index).head(5)
df.loc[df['quantity']==3].head().reset_index(drop=True)
```

```python
# Q23. quantity, item_price 두 개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라 
df[['quantity', 'item_price']]
```

```python
# Q24. item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라
df['new_price'] = df['item_price'].str[1:].astype('float')
df['new_price'].head(5)
```

```python
# Q25. new_price 컬럼이 5 이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라 
# df[df['new_price']<=5].shape[0]
len(df[df['new_price']<=5])
```

```python
# Q26. item_name명이 Chicken Salad Bowl인 데이터 프레임을 추출하고 index 값을 초기화해라 
# Ans = df[df['item_name']=='Chicken Salad Bowl'].reset_index(drop=True)
Ans = df.loc[df.item_name =='Chicken Salad Bowl'].reset_index(drop=True)
Ans.head(3)
```

```python
# Q27. new_price 값이 9 이하고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라 
df[(df['new_price']<=9)&(df['item_name']=='Chicken Salad Bowl')].head(5)
```

```python
# Q28. df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화해라 
ans = df.sort_values(by='new_price').reset_index(drop=True)
ans.head(5)
```

```python
# Q29. df의 item_name 컬럼 값 중 Chips 포함하는 경우의 데이터를 출력하라
df[df['item_name'].str.contains('Chips')].head(5)
```

```python
# Q30. df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라
df.iloc[:,::2]
```

```python
# Q31. df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화해라
df.sort_values(by='new_price', ascending=False).reset_index(drop=True).head(5)
```

```python
# Q32. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱해라 
df[(df['item_name']=='Steak Salad')|(df['item_name']=='Bowl')].head(5)
```

```python
# Q33. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 데이터프레임화한 후, 
# item_name을 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라 
ans = df[(df['item_name']=='Steak Salad')|(df['item_name']=='Bowl')]
ans = ans.drop_duplicates('item_name')
ans
```

```python
# Q33. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 데이터프레임화한 후, 
# item_name을 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라 
ans = df[(df['item_name']=='Steak Salad')|(df['item_name']=='Bowl')]
ans = ans.drop_duplicates('item_name', keep='last')
ans
```

```python
# Q35. df의 데이터 중 new_price 값이 new_price 값의 평균값 이상을 가지는 데이터들을 인덱싱하라 
df[df['new_price']>=df['new_price'].mean()].head(5)
```

```python
# Q36. df의 데이터 중 item_name의 값이 Izze 데이터를 Fizzy Lizzy로 수정하라
df.loc[df['item_name']=='Izze', 'item_name'] = 'Fizzy Lizzy'
df.head(5)
```

```python
# Q37. df의 데이터 중 choice_description 값이 NaN인 데이터의 갯수를 구해라 
df['choice_description'].isnull().sum()
```

```python
# Q38. df의 데이터 중 choice_description 값이 NaN인 데이터를 NoData 값으로 대체하라(loc 이용)
df.loc[df['choice_description'].isnull(), 'choice_description'] = 'NoData'
df.head()
```

```python
# Q39. df의 데이터 중 choice_description 값에 Black이 들어가는 경우를 인덱싱하라 
df[df['choice_description'].str.contains('Black')].head()
```
